```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(50,1)

y = 1.3*(X**8) - 1.4*(X**7) - 0.8*(X**6) + 0.1* \
    np.random.rand(50, 1)*np.random.choice([-1, 1])
y_pred = 1.3*(X**8) - 1.4*(X**7) - 0.8*(X**6) +0.05

fig = plt.figure()
plt.scatter(X, y)
plt.scatter(X, y_pred)
plt.show()
```
